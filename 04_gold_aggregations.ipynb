```python
# Gold table 1 — daily revenue by region & channel
spark.sql("""
CREATE OR REPLACE TABLE gold_daily_revenue AS
SELECT
    order_date,
    region,
    channel,
    COUNT(DISTINCT order_id)   AS order_count,
    COUNT(DISTINCT customer_id) AS unique_customers,
    SUM(quantity)               AS units_sold,
    ROUND(SUM(revenue), 2)      AS total_revenue
FROM silver_sales
GROUP BY order_date, region, channel
""")

# Gold table 2 — product performance ranking
spark.sql("""
CREATE OR REPLACE TABLE gold_product_performance AS
SELECT
    product,
    region,
    ROUND(SUM(revenue), 2)          AS total_revenue,
    SUM(quantity)                    AS total_units,
    ROUND(AVG(unit_price), 2)        AS avg_price,
    RANK() OVER (ORDER BY SUM(revenue) DESC) AS revenue_rank
FROM silver_sales
GROUP BY product, region
""")

# Optimise for DirectLake
spark.sql("OPTIMIZE gold_daily_revenue")
spark.sql("OPTIMIZE gold_product_performance")
print("Gold tables ready")
```
