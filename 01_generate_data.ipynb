```python
import pandas as pd, numpy as np
from datetime import datetime, timedelta

np.random.seed(42)
n = 50000

products = ['Laptop','Phone','Tablet','Monitor','Keyboard','Mouse','Headphones','Webcam']
regions  = ['North','South','East','West']
channels = ['Online','In-Store','Mobile']

df = pd.DataFrame({
    'order_id':    [f'ORD-{i:06d}' for i in range(n)],
    'order_date':  [datetime(2024,1,1) + timedelta(days=int(d))
                    for d in np.random.randint(0, 365, n)],
    'product':     np.random.choice(products, n),
    'region':      np.random.choice(regions, n),
    'channel':     np.random.choice(channels, n),
    'quantity':    np.random.randint(1, 10, n),
    'unit_price':  np.round(np.random.uniform(10, 2000, n), 2),
    'customer_id': [f'CUST-{np.random.randint(1,5000):05d}' for _ in range(n)],
})

df['revenue'] = df['quantity'] * df['unit_price']

# Introduce realistic messiness
df.loc[np.random.choice(df.index, 500), 'region'] = None
df.loc[np.random.choice(df.index, 200), 'unit_price'] = -1

spark_df = spark.createDataFrame(df)
spark_df.write.mode('overwrite').parquet('Files/raw/sales/')
print(f'Written {n:,} rows to Files/raw/sales/')
```
